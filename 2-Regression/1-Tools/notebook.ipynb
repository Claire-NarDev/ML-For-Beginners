```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
# On cherche combien le patient est malade en fonction de son BMI

# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

# Print the shape of the data and the first row

# (combien on a de points de data, combien d'attribut on a par point)
# soit (samples, features)
print(X.shape) 

print(X[0, :]) # Tous les attributs pour le premier patient
```

```python
# La feature BMI (?) est à l'index 2, on veut afficher tous les BMI
BMI = X[:, 2]
print(BMI.shape) # 1-D array

# Reshape to a 2-D array: on veut 442 arrays (sample) de taille 1 (features)
BMI = BMI.reshape((-1, 1)) # On veut (-1 = BMI.length) arrays de 1 colonne
print(BMI.shape)
```

```python
# Split the data into training (70%) and testing (30%) data
# y c'est le label (ici c'est combien il est malade)
X_train, X_test, y_train, y_test = model_selection.train_test_split(BMI, y, test_size=0.33)
```

```python
# Create a linear regression model and train it with our data
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict the disease progress (y_pred) with the BMI samples (X_test)
y_pred = model.predict(X_test)
```

```python
# Maintenant, on va tester à quel point notre modèle est bien ou pas

# Create a scatter plot for test values (avec plein de points)
plt.scatter(X_test, y_test, color='black')

# Plot the predictions
plt.plot(X_test, y_pred, color='blue', linewidth=3)

# Add labels and title
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')

# Draw the plot
plt.show()
```
