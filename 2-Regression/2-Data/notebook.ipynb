```python
import pandas as pd

# Read the US pumpkin CSV file
pumpkins = pd.read_csv('../data/US-pumpkins.csv')

# Filter out any rows that don't use the price per bushel
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case=True, regex=True)]
pumpkins.head()
```

```python
# Count how many cells in each column are empty
pumpkins.isnull().sum()
```

```python
# Define the columns we want to keep
new_columns = ['Package', 'Month', 'Low Price', 'High Price', 'Date']
light_pumpkins = pumpkins.drop([c for c in pumpkins.columns if c not in new_columns], axis=1)
```

```python
# Calculate the average price from the high and low price columns
price = (light_pumpkins['Low Price'] + light_pumpkins['High Price']) / 2

# Get the month from the date column
month = pd.DatetimeIndex(light_pumpkins['Date']).month

# Create a new data frame with these columns
new_pumpkins = pd.DataFrame({
    'Month': month, 
    'Package': light_pumpkins['Package'],
    'Low Price': light_pumpkins['Low Price'],
    'High Price': light_pumpkins['High Price']
                            })
```

```python
# Convert the price of all cells prices by 1 1/9 bushels by dividing by 1 1/9
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'), 'Price'] = price/(1 + 1/9)

# Convert the price of all cells prices by 1/2 bushels by dividing by 1/2
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1/2'), 'Price'] = price/(1/2)
```

```python
# Print the first rows from data frame
new_pumpkins.head()
```

```python
import matplotlib.pyplot as plt

# Get the values we want to plot
price = new_pumpkins.Price
month = new_pumpkins.Month

# Create and show a scatter plot of price vs month
plt.scatter(price, month)
plt.show()
```

Avec le graph ci-dessus, on ne peut pas réellement répondre à la question: Quel est le mois où l'on peut acheter des courges le moins cher ?

On essaie donc autre chose. On va afficher le prix moyen par mois avec un bar chart.

```python
# Group the pumpkins into a bar chart by month and price
new_pumpkins.groupby(['Month'])['Price'].mean().plot(kind='bar')

# Add a label to the bar chart
plt.ylabel("Pumpkin Price")
```
